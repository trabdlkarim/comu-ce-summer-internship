# Finding the top 15 infected countries

```python
def find_top_confirmed(dataset,n = 15):
    countries = dataset.groupby('Country_Region').sum()[['Confirmed', 'Deaths', 'Recovered', 'Active']]
    cdf = countries.nlargest(n, 'Confirmed')[['Confirmed']]
    return cdf
```

```python
cdf = find_top_confirmed(dataset2)
cdf
```

```python
confirmed_cases = [(country,confirmed) for country,confirmed in zip(cdf.index,cdf['Confirmed'])]
```

```python
confirmed_cases
```

```python
data = dataset2[['Lat','Long_','Confirmed']]
data = data.dropna()
```

```python
def circle_maker(x,m):
    flm.Circle(location=[x[0],x[1]], radius=float(x[2]), color="red",
                  popup='confirmed cases:{}'.format(x[2])).add_to(m)
```

```python
data.apply(lambda x:circle_maker(x,covidmap),axis=1)
```
